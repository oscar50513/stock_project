```python
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates

SData = open('2886_兆豐金1.csv').readlines()
SData1 = [line.strip('\n').split(',') for line in SData[1:] ]
s_date = [datetime.strptime(line[0], '%Y/%m/%d') for line in SData1]

s_close = [float(line[4]) for line in SData1]
s_volume = [float(line[5]) for line in SData1]

plt.figure(figsize = (12,6))

plt.subplot(211)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y/%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator()) 
plt.grid()
plt.plot(s_date, s_close)


plt.subplot(212)
plt.bar(s_date, s_volume, color ='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates

SData = open('0050_台灣50.csv').readlines()
SData1 = [line.strip('\n').split(',') for line in SData[1:] ]
s_date = [datetime.strptime(line[0], '%Y/%m/%d') for line in SData1]
s_close = [float(line[2]) for line in SData1]
s_close_avg = []
s_volume_avg = []
s_date2 = []



for y in range(2017,2020):
    a = [line for line in SData1 if datetime.strptime(line[0], '%Y/%m/%d').year == y]
    for m in range(1,13):
        s_date2.append(str(y)+"/"+str(m))
        close = 0
        num = 0
        close_avg = 0
        volume = 0
        SData2 = [line for line in a if datetime.strptime(line[0], '%Y/%m/%d').month == m]
        s_volume = [float(line[3]) for line in SData2]
        for data in s_volume:
            volume = volume + data
        s_volume_avg.append(round(volume))
        s_close = [float(line[2]) for line in SData2]
        for close_data in s_close:
            close = close + close_data
            num = num + 1
        close_avg = close / num
        s_close_avg.append(round(close_avg))

    
plt.figure(figsize = (12,6))

plt.subplot(211)
plt.plot(s_date2, s_close_avg)


plt.subplot(212)
plt.bar(s_date2,s_volume_avg,color ='red')
plt.gcf().autofmt_xdate()

plt.show()
```
