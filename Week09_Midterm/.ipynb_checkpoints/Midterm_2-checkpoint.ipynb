```python
# 呂皇逸
import matplotlib.pyplot as plt
import pandas as pd
datas = pd.read_csv("2886.csv")

avg = []
dividend=[]
day =[]
sum = 0
count = []
counts = 0
for lines in datas['dividend']:
    dividend.append(lines)
    count.append(counts)
    sum+=lines
    counts = counts+1

sum = round((sum /19),2)
sum1 = 0
for lines in datas['day']:
    day.append(lines)
    sum1+=lines
sum1 = round((sum1 /19),2)
for x in count:
    avg.append(dividend[x]/day[x])
plt.figure(figsize=(15,12))
plt.bar(datas.year,datas.dividend,color='orange',label=sum)
plt.legend()
plt.xlabel("year")
plt.ylabel("Price")
plt.title("dividend")
plt.show()
plt.figure(figsize=(15,12))
plt.plot(datas.year,datas.day,"r-o",label=sum1)
plt.legend()
plt.xlabel("year")
plt.ylabel("day")
plt.title("ex-dividend day")
```

```python
# 410921229 
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
Data = open('2886.csv').readlines()
SData = [line.strip('n').split(',') for line in Data[1:] ]
year = [datetime.strptime(i[0],'%Y') for i in SData]
dividend = [float(i[1]) for i in SData]
ex_dividend_days = [float(i[2]) for i in SData]

plt.figure(figsize = (8,12))
plt.subplot(211)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.bar(year,dividend,width=300,color='orange',label='1.25')
plt.legend()
plt.xlabel('year')
plt.ylabel('price')
plt.grid(0)
plt.title('dividend')
plt.xticks(rotation=30)

plt.subplot(212)
plt.plot(year,ex_dividend_days,'r-',marker='o',label='83.37')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.xticks(rotation=30)
plt.grid(0)
plt.legend(loc=2)
plt.title('ex-dividend days')
plt.xlabel('year')
plt.ylabel('day')
```

```python
#  410721202 
import matplotlib.pyplot as plt
import numpy as np
data = open('2886.csv').readlines()
data = [line.strip('n').split(',') for line in data[1:]]
year = [str(line[0]) for line in data]
price = [float(line[1]) for line in data]
day = [int(line[2]) for line in data]

mp = round(np.array(price).mean(), 2)
md = round(np.array(day).mean(), 2)

fig = plt.figure(figsize=(12, 8))
plt.title('dividend')
plt.xlabel('year')
plt.ylabel('price')

plt.bar(year, price, color='orange', label=mp)
plt.yticks(np.arange(0, 1.7, 0.2))
plt.legend(loc='upper left')
plt.show()

fig = plt.figure(figsize=(12, 8))
plt.title('ex-dividend_days')
plt.xlabel('year')
plt.ylabel('day')
plt.yticks(np.arange(0, 251, 50))
plt.plot(year, day, color='r', marker='o', label=md)
plt.legend(loc='upper left')
plt.show()
```

```python
# 劉沛妮 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
SData1 = open('2886.csv').readlines()
SData1 = [line.strip('n').split(',') for line in SData1[1:]]
year = [line[0] for line in SData1]
price = [float(line[1]) for line in SData1]
day = [int(line[2]) for line in SData1]
sump = 0
n = 0
sumd = 0
for i in price:
    sump =sump+i
    n+=1
pm = round(sump/n,2)
for i in day:
    sumd =sumd+i
dm = round(sumd/n,2)
fig = plt.figure(figsize=(12,8))
plt.subplot(211)
plt.bar(year, price, label=str(pm), color ='orange')
plt.xticks(rotation=20)
plt.title("dividend")
plt.xlabel("year")
plt.ylabel("price")
plt.legend()
plt.subplot(212)
plt.plot(year, day, label=str(dm), color ='r', marker ='o')
plt.xticks(rotation=20)
plt.title("ex-dividend days")
plt.xlabel("year")
plt.ylabel("day")
plt.legend(loc=2)
plt.subplots_adjust(hspace=0.5)
```

```python
# 謝旻倫
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
import numpy as np

data2886 = open('2886.csv').readlines()
data2886_1 = [line.strip('n').split(',') for line in data2886[1:]]
year = [line[0] for line in data2886_1]
dividend = [float(line[1]) for line in data2886_1]
pmean = 0
for i in dividend:
    pmean += i 
day = [int(line[2]) for line in data2886_1]
pmean = round(pmean/ len(year),2)
daymean = 0
for i in day:
    daymean+=i
daymean = round(daymean/ len(year),2)
fig = plt.figure(figsize = (8,12))
ax1 = fig.add_subplot(211)
ax1.bar(year, dividend, label = pmean)
ax1.set_title('dividend')
ax1.set_ylabel('price')
ax1.set_xlabel('year')
ax1.legend()
ax2 = fig.add_subplot(212)
ax2.plot(year , day, marker = 'o', label = daymean, color = 'r')
ax2.set_title('ex-dividend_days')
ax2.set_ylabel('day')
ax2.set_xlabel('year')
ax2.legend(loc = 2)
plt.gcf().autofmt_xdate()
plt.show()
```

```python
#  熊亭媛 
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates

SData = open("2886.csv").readlines()
SData1 = [line.strip('n').split(',')for line in SData[1:]]

s_date = [str(line[0]) for line in SData1]
s_div = [float(line[1]) for line in SData1]
s_day = [float(line[2]) for line in SData1]
print(s_date)
d =0
m = 0
t = 0
for i in s_div:
    m = m + i
    t =t+1
m = round(m/t,2)
t = 0
for i in s_day:
    d = d + i
    t =t+1
    
d = round(d/t,2)
print(d)

plt.figure(figsize = (8,12))

plt.subplot(211)
plt.title("dividend")
plt.xlabel("year")
plt.ylabel("price")
plt.bar(s_date,s_div,label = str(m))
plt.legend()

plt.subplot(212)
plt.title("ex-dividend days")
plt.xlabel("year")
plt.ylabel("day")
plt.plot(s_date,s_day,'r.-',label = str(d))
plt.legend(loc = "upper left")
```

```python
#  楊秉樺
import matplotlib.pyplot as plt
import pandas as pd
sdata = pd.read_csv('2886.csv')
s_year = [str(line) for line in sdata['year']]
s_dividend = [line for line in sdata['dividend']]
s_day = [line for line in sdata['day']]
me_dividend = round((sum(s_dividend) / len(s_dividend)), 2)
me_day = round((sum(s_day) / len(s_day)), 2)
plt.figure(figsize = (12, 8))
#p1
plt.subplot(211)
plt.grid(0)
plt.bar(s_year, s_dividend, color = 'orange', label = me_dividend)
plt.legend(loc = 2)
plt.ylim((0, 1.6))
plt.title('dividend')
plt.xlabel('year')
plt.ylabel('price')
plt.gcf().autofmt_xdate()

#p2
plt.subplot(212)
plt.grid(0)
plt.plot(s_year, s_day, 'r', label = me_day)
plt.legend(loc = 2)
plt.title('ex-dividend days')
plt.xlabel('year')
plt.ylabel('day')
plt.gcf().autofmt_xdate()
plt.show()
```
