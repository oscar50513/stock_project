```python
import os
path = os.getcwd()
print(path)
lists = os.listdir(path)
print(lists[:4])
files = [file for file in lists if '.csv' in file]
print(files[:4])
print(len(files))
```

```python
for file in files:
    print(file)
    with open(file,'r',encoding='utf8') as data:
        lines = data.readlines()[1:]
        #print(lines[:5])
        year = lines[0][:4]
        #print(type(year))
        head = 'Date,Open,High,Low,Close,Volume\n'
        need = []
        need.append(head)
        for i in lines:
            need.append(i)
        #print(need[:5])
    with open('newdata/'+ file[:-4] +'_'+year+'.csv','w',encoding ='utf8') as f:
        f.writelines(need)
    print('finish')
```
