```python
# plt.xticks(x, label, rotation)
# rotation 設定旋轉角度!
import numpy as np
import matplotlib.pyplot as plt
name = ['A', 'B', 'C']
En = [80,60,55]
Ma = [77,80,99]
x = np.arange(len(En))
print('x:', x)

plt.figure(figsize=(10,5))
plt.plot(x, En, '^-', label = 'En')
plt.plot(x, Ma, 'o-', label = 'Ma')
plt.legend() # 印出 label
plt.grid()
plt.title('origin')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(x, En, '^-', label = 'En')
plt.plot(x, Ma, 'o-', label = 'Ma')
plt.legend()
plt.grid()
plt.title('use xticks')
plt.xticks(x,name, rotation=0)
```

```python
# 繪製畫框軸
# 左下的XY座標 到 右上的XY座標

fig = plt.figure(figsize=(8,4))
ax1 = fig.add_axes([0,0.3,1,1])
ax2 = fig.add_axes([0,0,1,0.3])

# 設定坐標軸名稱

ax1.set(ylabel = "ax1_Y", title = "test")
ax2.set(ylabel = "ax2_Y", xlabel = "ax2_X")
```

```python
# 繪製直線: np.full(個數, 數值)
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(20)
y = x+1

line_up = np.full(len(x), 15)
line_bt = np.full(len(x), 5)

plt.plot(x,y)
plt.plot(x,line_up, 'r')
plt.plot(x,line_bt, 'g')
plt.show()
```
